```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate random numbers from three different distributions
data1 = np.random.normal(loc=700, scale=150, size=1000)
data2 = np.random.normal(loc=100, scale=10, size=100)
data3 = np.random.randint(low=0, high=1000, size=500)

# Combine the data into a single array
data = np.concatenate((data1, data2, data3))

# Create a DataFrame from the data
df = pd.DataFrame({'TTL': data})

# Plot histogram with 20 bins
plt.hist(df['TTL'], bins=20)
plt.xlabel('TTL')
plt.ylabel('Frequency')
plt.title('Bimodal Distribution Histogram')

# Show the plot
plt.show()
```

```python
hist, bin_edges = np.histogram(data, bins=100, range=(0, 1000))
cdf = np.cumsum(hist) / np.sum(hist)
df_cdf = pd.DataFrame({'Value': bin_edges[:-1], 'CDF': cdf})
```

```python
# Plot the CDF
plt.plot(df_cdf['Value'], df_cdf['CDF'])
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Bimodal Distribution CDF')

# Show the plot
plt.show()
```

```python
# Make a new dataset
data_comparison = np.concatenate(
    (
        np.random.normal(loc=700, scale=100, size=1000),
        np.random.normal(loc=50, scale=10, size=100),
        np.random.randint(low=0, high=1000, size=500)
    )
)

# Create a DataFrame from the data
df_comparison = pd.DataFrame({'TTL': data_comparison})

# Plot histogram with 20 bins
plt.hist(df_comparison['TTL'], bins=20)
plt.xlabel('TTL')
plt.ylabel('Frequency')
plt.title('Bimodal Distribution Histogram')

# Show the plot
plt.show()
```

```python
hist, bin_edges = np.histogram(data_comparison, bins=100, range=(0, 1000))
cdf = np.cumsum(hist) / np.sum(hist)
df_cdf_comparison = pd.DataFrame({'Value': bin_edges[:-1], 'CDF': cdf})
```

```python
# Plot the CDF
plt.plot(df_cdf_comparison['Value'], df_cdf_comparison['CDF'])
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Bimodal Distribution CDF')

# Show the plot
plt.show()
```

```python
# Plot both of the CDFs
plt.plot(df_cdf_comparison['Value'], df_cdf_comparison['CDF'])
plt.plot(df_cdf['Value'], df_cdf['CDF'])
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Bimodal Distribution CDF')

# Show the plot
plt.show()
```
